```python
A1 = ['Junior consultant ','Consultant ','Senior consultant ','Lead consultant ','Manager ','Contractor ']
A2 = ['information ','solution ','help ']
A3 = ['call','service','consultation','advice','connection']

B1 = ['Rozmowa ','Usługa ','Konsultancja ','Porada ','Połącznie ']
B2 = ['informacyjna ','ws rozwiązania ','ws pomocy ']
B3 = ['młodszego konsultanta','konsultanta','starszego konsultanta','głównego konsultanta','manadżera','kontraktora']
```

```python
# Cartesian product of service name both in Polish and English

A = list()
B = list()

for i in itertools.product(A1,A2,A3):
    A.append(i)

for i in itertools.product(B1,B2,B3):
    B.append(i)

df = pd.DataFrame({
    "serviceFullNameEng" : [a[0] + a[1] + a[2] for a in A],
    "serviceFullNamePl"  : [b[0] + b[1] + b[2] for b in B]
})

df['serviceShortNameEng'] = [ i[0:4] + j[0:4].title() + k[0:4].title() for i,j,k in A]
df['serviceShortNamePl']  = [ i[0:4] + j[0:4].title() + k[0:4].title() for i,j,k in B]

df['serviceCode'] = [ i[0:6].upper() + j[0:4].upper() + k[0:4].upper() + '00' for i,j,k in A]

sfnp = ('000' + (df.serviceFullNamePl.str.len()).astype(str))
df['sfnp'] = [s[-4:] for s in sfnp]

# Create service business key
df['serviceBusinessKey'] = [ i[0:1].upper() + j[0:1].upper() + k[0:4].upper() for i,j,k in A] + df.sfnp

df = df.drop(['sfnp'], axis = 1)

df
```

```python
# Create csv file with given destination

df.columns = column_list
df = df.sample(n = obs, random_state = 123)
df.to_csv(path_to_file, index = None, header = True)
```
