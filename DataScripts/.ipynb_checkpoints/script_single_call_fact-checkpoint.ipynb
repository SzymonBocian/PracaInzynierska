```python

l_conn_dict = { 
    "user"     : "szymonbocian", \
    "password" : "", \
    "host"     : "localhost", \
    "port"     : "5432", \
    "database" : "dwh_call_center"
}

l_obs_dict = {
    "min" : 10000, \
    "max" : 50000
}

l_week_dict = {
    "year" : 2019, \
    "week" : 10
}

l_time_dict = {
    "start" : "08:00:00", \
    "end"   : "22:00:00"
}

# end call after 1 hour of waiting - 60 sec x 60 min
# min 5 minutes of conversation - 60 sec x 5 min
l_join_dict = {
    "wait_min" : 10, \
    "wait_max" : 1800, \
    "conv_min" : 300, \
    "conv_max" : 3600
}

l_skill_dict = pd.read_csv( skill_dict_path,
                            sep = ";",
                            usecols = ['skill_key','skill_name'],
                            skiprows = 0
                          )

l_obs = randrange(l_obs_dict["min"],l_obs_dict["max"]) # number of observation
```

```python
# Create calendar as cartesian product of date x hours from Postgres

l_conn = pm.DbConn(l_conn_dict)

l_query_list = []

l_query_list.append("""
    SELECT 1, date_key, date_value FROM tech.date_dim 
    WHERE year_value = %(year)d AND iso_week_of_year = %(week)d;""" % l_week_dict)

l_query_list.append("""
    SELECT 1, time_key, time_value FROM tech.time_dim 
    WHERE time_value >= '%(start)s' AND time_value <= '%(end)s';""" % l_time_dict)

l_query_list.append("""
    SELECT office_key, office_business_key FROM pro.office_dim;""")

l_date_df = pd.DataFrame(l_conn.run_sql(l_query_list[0]), columns = ["key", "date_key", "date_value"])
l_time_df = pd.DataFrame(l_conn.run_sql(l_query_list[1]), columns = ["key", "time_key", "time_value"])

l_office_df = pd.DataFrame(l_conn.run_sql(l_query_list[2]), columns = ["office_key", "office_business_key"])

# full calendar start_week - end_week ; start_hour - end_hour
l_calendar_df = pd.merge(l_date_df, l_time_df, on = "key")[["date_key", "date_value", "time_key", "time_value"]]
```

```python

l_single_call_df = pd.DataFrame({
    "client_business_key"     : ["SB048048025M"] * l_obs, # temporary 
    "consultant_business_key" : ["AA032048000E"] * l_obs, # temporary
    "office_business_key"     : np.random.choice(l_office_df["office_business_key"], l_obs),
    "service_key"             : np.random.choice(l_skill_dict["skill_name"], l_obs), # re-work is needed
    "start_join_date_key"     : l_calendar_df.date_key.sample(n = l_obs)
})

l_single_call_df = pd.merge( l_single_call_df, 
                             l_calendar_df, 
                             left_index = True, 
                             right_index = True)[["client_business_key","consultant_business_key","office_business_key","service_key","date_key","date_value","time_key","time_value"]]

l_single_call_df["start_join_date_value"] = pd.to_datetime(l_single_call_df["date_value"])
l_single_call_df["start_join_time_value"] = pd.to_timedelta(l_single_call_df["time_value"].astype(str))

l_single_call_df["wait_time"] = [ randrange(l_join_dict["wait_min"],l_join_dict["wait_max"]) for l in l_single_call_df.index ]
l_single_call_df["wait_time"] = pd.to_timedelta(l_single_call_df["wait_time"], unit = "s")

l_single_call_df["start_conv_date_value"] = l_single_call_df["start_join_date_value"]
l_single_call_df["start_conv_time_value"] = l_single_call_df["start_join_time_value"] + l_single_call_df["wait_time"]

l_single_call_df["conv_time"] = [ randrange(l_join_dict["conv_min"],l_join_dict["conv_max"]) for l in l_single_call_df.index ]
l_single_call_df["conv_time"] = pd.to_timedelta(l_single_call_df["conv_time"], unit = "s")

l_single_call_df["end_conn_date_value"] = l_single_call_df["start_conv_date_value"]
l_single_call_df["end_conn_time_value"] = l_single_call_df["start_conv_time_value"] + l_single_call_df["conv_time"]

l_single_call_df["start_join_timestamp"] = l_single_call_df["start_join_date_value"] + l_single_call_df["start_join_time_value"]

l_single_call_df = l_single_call_df.drop(["date_key","date_value","time_key","time_value","wait_time","conv_time"], axis = 1)

l_single_call_df.head(10)
```

```python
l_single_call_df.to_csv(output_path, 
                        index = None, 
                        header = True)
```
