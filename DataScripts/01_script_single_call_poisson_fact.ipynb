```python


l_poisson_distribution = get_poisson_distribution(4.75, 10000)

count, bins, ignored = plt.hist(l_poisson_distribution, 14, density=False)
plt.show()
print(max(l_poisson_distribution), min(l_poisson_distribution))

def get_random_clock_value(l_bound=0, h_bound=59):
    return ":" + str(randrange(0, 59)).zfill(2)

def get_poisson_distribution(l_lam, l_obs):
    l_max = 0
    while(l_max != 15):
        l_poisson_distribution = np.random.poisson(lam=(l_lam), size=(l_obs, 1))
        l_max = max(l_poisson_distribution)
    return l_poisson_distribution
```

```python

l_output_df = pd.DataFrame(columns = ["client_business_key", "consultant_business_key", 
                                      "office_business_key", "service_business_key", 
                                      "start_join_date_value", "start_conv_date_value",
                                      "end_conn_date_value", "start_join_time_value",
                                      "start_conv_time_value", "end_conn_time_value"
                                     ])

for index, day in l_calendar_df.iterrows():
    l_obs = randrange(l_obs_dict["min"],l_obs_dict["max"]) # number of observation
#     l_obs = randrange(1,10)
    
    # get Poisson distribution with l_obs observations
    l_poisson_distr = get_poisson_distribution(4.75, l_obs)
    # attach Poisson distribution to working hour
    l_poisson_distr = l_poisson_distr + int(l_period_dict["start"][:2])
    # convert ndarray integer values to string values
    l_tmp_p_distr = [",".join(p).zfill(2) for p in l_poisson_distr.astype(str)]
    # generate random time of calling based on Poisson distribuation
    l_p_distr = [p + get_random_clock_value() + get_random_clock_value() for p in l_tmp_p_distr]
    
    l_tmp_df = pd.DataFrame({
        "client_business_key"     : ["SB048048025M"] * l_obs, # temporary 
        "consultant_business_key" : ["AA032048000E"] * l_obs, # temporary
        "office_business_key"     : np.random.choice(l_office_df["office_business_key"], l_obs),
        "service_business_key"    : np.random.choice(l_service_df["service_business_key"], l_obs),
        "start_join_date_value"   : day["date_value"],
        "start_conv_date_value"   : day["date_value"],
        "end_conn_date_value"     : day["date_value"],
        "start_join_time_value"   : pd.to_timedelta(l_p_distr)
    })
    l_tmp_df["start_conv_time_value"] = l_tmp_df["start_join_time_value"] \
        .apply(lambda x: \
               x + pd.to_timedelta(randrange(l_join_dict["wait_min"],l_join_dict["wait_max"]), unit = "s"))
    
    l_tmp_df["end_conn_time_value"] = l_tmp_df["start_conv_time_value"] \
        .apply(lambda x: \
               x + pd.to_timedelta(randrange(l_join_dict["conv_min"],l_join_dict["conv_max"]), unit = "s"))
                                                                                
    l_output_df = pd.concat([l_output_df, l_tmp_df])
    
l_output_df.head(1000)
```
